```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import GaussianNB
```

```python
dataset=pd.read_csv("heart.csv")
dataset=dataset.fillna(0)
dataset.head()
```

```python
labelData=dataset['target']
del dataset['target']
features=dataset.values#vector de caracteristicas
labels=labelData.values#vector de etiquetas/clases
```

```python
import pandas
from sklearn.model_selection import KFold
import numpy as np
X = features
y = labels
```

```python
cv = KFold(n_splits=10, random_state=0, shuffle=False)
print(cv)
gnb = GaussianNB()
scores = []
for train_index, test_index in cv.split(X):
    X_trainc, X_testc, y_trainc, y_testc = X[train_index], X[test_index], y[train_index], y[test_index]
    gnb.fit(X_trainc,y_trainc)
    predictions=gnb.predict(X_testc)
    scores.append(gnb.score(X_testc, y_testc))
    print(confusion_matrix(y_testc,predictions))
    print(classification_report(y_testc,predictions))
    print(accuracy_score(y_testc,predictions))
```

```python
print("Acuaraccy: ",np.mean(scores)*100,"%")
```
